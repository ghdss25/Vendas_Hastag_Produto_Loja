## Analise de Vendas do Mês de Dezembro

```python
valor_unitario_total = vendas_dez["Valor Unitário"].sum()
valor_final_total = vendas_dez["Valor Final"].sum()

valor_unitario_medio = vendas_dez["Valor Unitário"].mean() 
valor_final_medio = vendas_dez["Valor Final"].mean() 

codigo_venda = len(vendas_dez["Código Venda"].unique())

quantidade = vendas_dez["Quantidade"].count()

analise_venda = pd.DataFrame({"Valor Total Unitário" : [valor_unitario_total], 
                              "Valor Final": [valor_final_total], 
                              "Média Unitária" : [valor_unitario_medio], 
                              "Média Final" : [valor_final_medio], 
                              "Número de Códigos Venda" : [codigo_venda], 
                              "Quantidade de Produtos" : [quantidade]})

analise_venda["Valor Total Unitário"] = analise_venda["Valor Total Unitário"].map("R$ {:,.2f}".format)
analise_venda["Valor Final"] = analise_venda["Valor Final"].map("R$ {:,.2f}".format)
analise_venda["Média Unitária"] = analise_venda["Média Unitária"].map("R$ {:,.2f}".format)
analise_venda["Média Final"] = analise_venda["Média Final"].map(" R$ {:,.2f}".format)

analise_venda
```

## Analise geral das Lojas de Vendas do Mês de Dezembro por Valor Unitário

```python
loja_total_unitario = vendas_dez.groupby(["Nome Loja"]).sum()["Valor Unitário"].rename("Total Unitária") 
loja_media_unitario = vendas_dez.groupby(["Nome Loja"]).mean()["Valor Unitário"].rename("Média Unitária")
loja_mediana_unitario = vendas_dez.groupby(["Nome Loja"]).median()["Valor Unitário"].rename("Mediana Unitária") 
loja_desvio_unitario = vendas_dez.groupby(["Nome Loja"]).std()["Valor Unitário"].rename("Desvio Padrão") 
loja_variancia_unitario = vendas_dez.groupby(["Nome Loja"]).var()["Valor Unitário"].rename("Variância")


analise_total_loja = pd.DataFrame({"Total Unitária" : loja_total_unitario, 
                                   "Média Unitária" : loja_media_unitario, 
                                   "Mediana Unitária" : loja_mediana_unitario,
                                   "Desvio Padrão" : loja_desvio_unitario, 
                                   "Variância" : loja_variancia_unitario})

analise_total_loja["Total Unitária"] = analise_total_loja["Total Unitária"].map("R$ {:,.2f}".format)
analise_total_loja["Média Unitária"] = analise_total_loja["Média Unitária"].map("R$ {:,.2f}".format)
analise_total_loja["Mediana Unitária"] = analise_total_loja["Mediana Unitária"].map("R$ {:,.2f}".format)
analise_total_loja["Desvio Padrão"] = analise_total_loja["Desvio Padrão"].map("R$ {:,.2f}".format) 
analise_total_loja["Variância"] = analise_total_loja["Variância"].map("R$ {:,.2f}".format) 
analise_total_loja
```

```python
moda_loja = pd.DataFrame() 

for n in list(vendas_dez["Nome Loja"].unique()): 
    
    vu = vendas_dez[(vendas_dez["Nome Loja"] == n)]["Valor Unitário"].mode() 
    moda_loja = moda_loja.append(pd.DataFrame({'Nome Loja' : n, 'Moda Unitária': vu}, index=[0]))
        
moda_loja = moda_loja.reset_index() 
del moda_loja["index"] 

moda_loja["Moda Unitária"].fillna(0, inplace=True) 
moda_loja["Moda Unitária"].isnull().value_counts()
moda_loja["Moda Unitária"] = moda_loja["Moda Unitária"].map("R$ {:,.2f}".format)

moda_loja
```

## Analise geral das Lojas de Vendas do Mês de Dezembro por Valor Final

```python
loja_total_final = vendas_dez.groupby(["Nome Loja"]).sum()["Valor Final"].rename("Total Final") 
loja_media_final = vendas_dez.groupby(["Nome Loja"]).mean()["Valor Final"].rename("Média Final") 
loja_mediana_final = vendas_dez.groupby(["Nome Loja"]).median()["Valor Final"].rename("Mediana Final") 
loja_desvio_final = vendas_dez.groupby(["Nome Loja"]).std()["Valor Final"].rename("Desvio Padrão") 
loja_variancia_final = vendas_dez.groupby(["Nome Loja"]).var()["Valor Final"].rename("Variância") 

analise_loja_final = pd.DataFrame({"Total Final" : loja_total_final, 
                                   "Média Final" : loja_media_final, 
                                   "Mediana Final" : loja_mediana_final, 
                                   "Desvio Padrão" : loja_desvio_final, 
                                   "Variância" : loja_variancia_final}) 

analise_loja_final["Total Final"] = analise_loja_final["Total Final"].map("R$ {:,.2f}".format)
analise_loja_final["Média Final"] = analise_loja_final["Média Final"].map("R$ {:,.2f}".format)
analise_loja_final["Mediana Final"] = analise_loja_final["Mediana Final"].map("R$ {:,.2f}".format) 
analise_loja_final["Desvio Padrão"] = analise_loja_final["Desvio Padrão"].map("R$ {:,.2f}".format)
analise_loja_final["Variância"] = analise_loja_final["Variância"].map("R$ {:,.2f}".format)
analise_loja_final

```

```python
moda_loja_final = pd.DataFrame() 

for n in list(vendas_dez["Nome Loja"].unique()): 
    
    vf = vendas_dez[(vendas_dez["Nome Loja"] == n)]["Valor Final"].mode() 
    moda_loja_final = moda_loja_final.append(pd.DataFrame({'Nome Loja' : n, 'Moda Unitária Final': vf}, index=[0]))
        
moda_loja_final = moda_loja_final.reset_index() 
del moda_loja_final["index"] 

moda_loja_final["Moda Unitária Final"].fillna(0, inplace=True) 
moda_loja_final["Moda Unitária Final"].isnull().value_counts()
moda_loja_final["Moda Unitária Final"] = moda_loja_final["Moda Unitária Final"].map("R$ {:,.2f}".format)

moda_loja_final
```

## Indice de Vendas por Nome da Loja 

```python
indice_vendas_loja = (loja_total_unitario / loja_total_final ) * 100
quantidade_lojas = vendas_dez.groupby(["Nome Loja"]).count()["Quantidade"].rename("Quantidade de Lojas")

analise_indice_vendas = pd.DataFrame({"Índice Vendas" : indice_vendas_loja, 
                                      "Quantidade de Lojas" : quantidade_lojas})

analise_indice_vendas["Índice Vendas"] = analise_indice_vendas["Índice Vendas"].map("{:,.1f}%".format)
analise_indice_vendas
```

## Analise Geral dos Produtos de Vendas do Mês de Dezembro por Valor Unitário

```python
produto_total_unitario = vendas_dez.groupby(["Produto"]).sum()["Valor Unitário"].rename("Total de Produtos")
produto_medio_unitario = vendas_dez.groupby(["Produto"]).mean()["Valor Unitário"].rename("Média de Produtos") 
produto_mediana_unitario = vendas_dez.groupby(["Produto"]).median()["Valor Unitário"].rename("Mediana de Produtos")
produto_desvio_unitario = vendas_dez.groupby(["Produto"]).std()["Valor Unitário"].rename("Desvio Padrão") 
produto_variancia_unitario = vendas_dez.groupby(["Produto"]).var()["Valor Unitário"].rename("Variância") 

analise_produto_geral = pd.DataFrame({"Total de Produtos Unitários" : produto_total_unitario,
                                      "Média de Produtos Unitários" : produto_medio_unitario, 
                                      "Mediana de Produtos Unitários" : produto_mediana_unitario, 
                                      "Desvio Padrão Unitários" : produto_desvio_unitario, 
                                      "Variância Unitárias" : produto_variancia_unitario})

analise_produto_geral["Total de Produtos Unitários"] = analise_produto_geral["Total de Produtos Unitários"].map("R$ {:,.2f}".format)
analise_produto_geral["Média de Produtos Unitários"] = analise_produto_geral["Média de Produtos Unitários"].map("R$ {:,.2f}".format)
analise_produto_geral["Mediana de Produtos Unitários"] = analise_produto_geral["Mediana de Produtos Unitários"].map("R$ {:,.2f}".format)
analise_produto_geral["Desvio Padrão Unitários"] = analise_produto_geral["Desvio Padrão Unitários"].map("R$ {:,.2f}".format) 
analise_produto_geral["Variância Unitárias"] = analise_produto_geral["Variância Unitárias"].map("R$ {:,.2f}".format)

analise_produto_geral.head(24)
```

```python
moda_produto_unitario = pd.DataFrame() 

for n in list(vendas_dez["Produto"].unique()):
    
    vu = vendas_dez[(vendas_dez["Produto"] == n)]["Valor Unitário"].mode() 
    moda_produto_unitario = moda_produto_unitario.append(pd.DataFrame({'Produto':n, 'Moda Unitária':vu}, index=[0]))

moda_produto_unitario = moda_produto_unitario.reset_index() 
del moda_produto_unitario["index"] 

moda_produto_unitario["Moda Unitária"].fillna(0, inplace=True) 
moda_produto_unitario["Moda Unitária"].isnull().value_counts()
moda_produto_unitario["Moda Unitária"] = moda_produto_unitario["Moda Unitária"].map("R$ {:,.2f}".format)
moda_produto_unitario.head(23)
```

## Analise Geral dos Produtos de Vendas do Mês de Dezembro por Valor Final

```python
produto_total_final = vendas_dez.groupby(["Produto"]).sum()["Valor Final"].rename("Preço Final do Produto") 
produto_medio_final = vendas_dez.groupby(["Produto"]).mean()["Valor Final"].rename("Preço Médio do Produto") 
produto_mediano_final = vendas_dez.groupby(["Produto"]).median()["Valor Final"].rename("Preço Mediano do Produto") 
produto_desvio_final = vendas_dez.groupby(["Produto"]).std()["Valor Final"].rename("Preço de Desvio Padrão") 
produto_variancia_final = vendas_dez.groupby(["Produto"]).var()["Valor Final"].rename("Variância de Preços Produto")

analise_produto_final = pd.DataFrame({"Preço Final do Produto" : produto_total_final, 
                                      "Preço Médio do Produto" : produto_medio_final, 
                                      "Preço Mediano do Produto" : produto_mediano_final, 
                                      "Preço de Desvio Padrão" : produto_desvio_final, 
                                      "Variância de Preços Produto" : produto_variancia_final}) 

analise_produto_final["Preço Final do Produto"] = analise_produto_final["Preço Final do Produto"].map("R$ {:,.2f}".format)
analise_produto_final["Preço Médio do Produto"] = analise_produto_final["Preço Médio do Produto"].map("R$ {:,.2f}".format)
analise_produto_final["Preço Mediano do Produto"] = analise_produto_final["Preço Mediano do Produto"].map("R$ {:,.2f}".format) 
analise_produto_final["Preço de Desvio Padrão"] = analise_produto_final["Preço de Desvio Padrão"].map("R$ {:,.2f}".format)
analise_produto_final["Variância de Preços Produto"] = analise_produto_final["Variância de Preços Produto"].map("R$ {:,.2f}".format) 

analise_produto_final.head(24)
```

```python
moda_produto_final = pd.DataFrame() 

for p in list(vendas_dez["Produto"].unique()): 
    
    pf = vendas_dez[(vendas_dez["Produto"] == p)]["Valor Final"].mode()
    moda_produto_final = moda_produto_final.append(pd.DataFrame({'Produto' : p, 'Moda Final' : pf}, index=[0])) 
    
moda_produto_final = moda_produto_final.reset_index() 

del moda_produto_final["index"] 
moda_produto_final["Moda Final"].fillna(0, inplace=True) 
moda_produto_final["Moda Final"].isnull().value_counts() 
moda_produto_final["Moda Final"] = moda_produto_final["Moda Final"].map("R$ {:,.2f}".format)

moda_produto_final.head(26)
```
